```python
import sqlite3
```

```python
con = sqlite3.connect('class_practice.db') # this is how we are going to create our database, 
                                         # calling it 'citibikeData.db'

# "con" stands for "connection" – this is telling SQLite what database to use
```

```python
import json 
import urllib.request # https://docs.python.org/3/library/urllib.request.html

with urllib.request.urlopen("https://data.cityofnewyork.us/resource/i4gi-tjb9.json") as url:
    data = json.loads(url.read().decode())
    print(data)
```

```python
# stations = data['stationBeanList'] # iterate through the json to find the station data of interest
```

```python
import pandas as pd # we'll use pandas just to visualize our data, NOT to query it

df_traffic = pd.DataFrame(data) # create a new dataframe called 'df_stations' 
df_traffic.head() # check the first five station entries
```

```python
# HINT: ignore 'encoded_poly_line', 'encoded_poly_line_lvls'
```

```python
sql = "CREATE TABLE IF NOT EXISTS TrafficData (traffic_id int, speed float, travel_time int, data_as_of datetime, link_id varchar(20), link_points varchar(250), owner varchar(250), transcom_id varchar(20), borough varchar(20), link_name varchar(250));" 

con.execute(sql)
con.commit()
```

```python
query_template = """INSERT OR IGNORE INTO TrafficData(traffic_id, speed, travel_time, data_as_of, link_id, \
link_points, owner, transcom_id, borough, link_name) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?);"""

for entry in data: # for every station entry in the json 
    traffic_id = entry['id'] # find and set station_id
    speed = float(entry['speed'])
    travel_time = int(entry['travel_time'])
    data_as_of = entry['data_as_of']
    link_id = str(entry['link_id'])
    link_points = str(entry['link_points']) 
    owner = str(entry['owner'])
    transcom_id = str(entry['transcom_id'])
    borough = str(entry['borough'])
    link_name = str(entry['link_name'])
                
    print("Inserting Traffic Data:", traffic_id, speed, travel_time, data_as_of, link_id, link_points, owner, transcom_id, borough, link_name) 
    
    query_parameters = (traffic_id, speed, travel_time, data_as_of, link_id, link_points, owner, transcom_id, borough, link_name) 
    
    con.execute(query_template, query_parameters)
    
con.commit()
```

Problem 2 (4 points):

> Use SELECT, DISTINCT, WHERE, and LIMIT in at least 4 _different_ queries to answer a question of your choice about our NYC traffic data. 

```python
# which borough has the most data points in our set? 

check = pd.read_sql("""SELECT borough, COUNT(borough) as borough_count
                       FROM TrafficData 
                       GROUP BY borough
                       """
                    , con=con)
check
```

```python
# what is the top speed reached in Brooklyn in our set? 

check = pd.read_sql("""SELECT speed
                       FROM TrafficData 
                       WHERE borough = 'Brooklyn'
                       ORDER BY speed DESC
                       LIMIT 1;
                       """
                    , con=con)
check
```

```python
# how many unique traffic id's do we have from the Bronx in our set? 

check = pd.read_sql("""SELECT COUNT(DISTINCT traffic_id) as unique_traffic_ids
                       FROM TrafficData 
                       WHERE borough = 'Bronx'
                       """
                    , con=con)
check
```

```python
# what are the top 5 longest travel times seen at WSE S ARDEN AVENUE - BLOOMINGDALE ROAD in our set?  

check = pd.read_sql("""SELECT link_name, travel_time
                       FROM TrafficData 
                       WHERE link_name = 'WSE S ARDEN AVENUE - BLOOMINGDALE ROAD'
                       ORDER BY travel_time DESC
                       LIMIT 5;
                       """
                    , con=con)
check
```
